```python
# ADDING COLUMNS

sales["global_category"] = sales["Global_Sales"].apply(lambda x:"minimum" if x < 20 else("medium" if x < 30 else "maximum"))
sales.head(30)


def category(row):
    if row["Global_Sales"] < 20:
        return"high"
    elif row["Global_Sales"] < 30:
        return "medium"
    else:
        return"low"
sales["global_sales2"] = sales.apply(category, axis = 1)
sales
```
